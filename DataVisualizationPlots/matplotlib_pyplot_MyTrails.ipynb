```python
import matplotlib.pyplot as plt
```

```python
#to show inline graph in jupyter notebook
%matplotlib inline
```

```python
import numpy as np
x= np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y) 
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Basic Graph')
# no need to call function plt.show()
```

```python
# subplot 
plt.subplot(1,2,1) # one row by two columns 
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# object orient methods 
figure = plt.figure() # plain canvas

axes = figure.add_axes([0.1,0.1,1,1])  # left, bottom, width, height (range 0 to 1)
#axes.plot(x,y) # This will update with new axis range
```

```python
# object orient methods 
figure = plt.figure() # plain canvas

axes = figure.add_axes([1,1,1,1])  # left, bottom, width, height (range 0 to 1)
#axes.plot(x,y) # This will update with new axis range
```

```python
twoplotfigures = plt.figure()
axes1 = twoplotfigures.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
axes1.set_title('Larger plot')
axes2 = twoplotfigures.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(y,x)
axes2.set_title('Smaller plot')
```

```python
#To bottom corner
bottomcornerfigures = plt.figure()
axes1= bottomcornerfigures.add_axes([0.1,0.1,1,1])
axes2= bottomcornerfigures.add_axes([0.5,0.2,0.3,0.3])
```

```python
fig,axescoll = plt.subplots(nrows=1,ncols=2)
#axescol1 is array of axes object
axescoll[0].plot(x,y)
axescoll[1].plot(y,x)
plt.tight_layout() # to make sure that sub plots are evenly spaced
```

```python
figsize = plt.figure(figsize=(5,2))
ax = figsize.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
figsize,figsizeaxescoll = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
figsizeaxescoll[0].plot(x,y)
figsizeaxescoll[1].plot(y,x)
plt.tight_layout()
```

```python
figsize.savefig('my_picture.png',dpi=200)
```

```python
legendfigure = plt.figure()
ax = legendfigure.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X squared')
ax.plot(x,x**3,label='X cubed')
ax.legend() # best = 0 , loc=10
```

```python
colorfigure = plt.figure()
coloraxes = colorfigure.add_axes([0,0,1,1])
coloraxes.plot(x,y,color='purple',linewidth=3,alpha=3,ls='--',
              marker='o',markerfacecolor='red',markeredgewidth=3,markeredgecolor='yellow')# RGB color codes, alpha is transparency 
```

```python
# limit the plot 
limitfigure = plt.figure(figsize=(6,3))
limitaxes = limitfigure.add_axes([0,0,1,1])
limitaxes.plot(x,y,c='purple',lw=2,ls='--')
limitaxes.set_xlim([0,1])
limitaxes.set_ylim([0,2])

```

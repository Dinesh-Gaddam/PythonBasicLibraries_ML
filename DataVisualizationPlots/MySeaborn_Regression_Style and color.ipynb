```python
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
tips = sb.load_dataset('tips')
tips.head()
```

```python

# Linear model plot
sb.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
# markers to o for male,v for female 
# To increase the size of vales scatter_kws
sb.lmplot(x='total_bill',y='tip',data=tips,hue='sex',
        markers=['o','v'], scatter_kws={'s':100})
```

```python
# coloumns instead of hue
sb.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
# row we use time where we get two more with lunch female  dinner female and so on
sb.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
# combine hue( sex) with cols( day ) and row (time)
sb.lmplot(x='total_bill',y='tip',data=tips, col='day',row='time'
         ,hue='sex',markers=['o','v'],scatter_kws={'s':100})
```

```python
# change aspect ratio 
sb.lmplot(x='total_bill',y='tip',data=tips, col='day',hue='sex',
          aspect=0.6,size=8)
```

```python
# styles and colours 
sb.set_style('whitegrid')
sb.countplot(x='sex',data=tips)
```

```python
# To have spines  
sb.set_style('ticks') # by default we have spines on the left/bottom
sb.countplot(x='sex',data=tips)
```

```python
# To have spines an the left
sb.set_style('ticks') 
sb.countplot(x='sex',data=tips)
sb.despine(left=True,bottom=True)
```

```python
# To use fig size
plt.figure(figsize=(12,3))
sb.countplot(x='sex',data=tips)
```

```python
# set context 
#plt.figure(figsize=(8,6))
# Its taking figure size set 
sb.set_context('poster',font_scale=1) # we can increase font size
sb.countplot(x='sex',data=tips)
```

```python
plt.figure(figsize=(10,8)) # default figure size still 
sb.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
# https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html#classes-of-colormaps
sb.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='seismic')
```

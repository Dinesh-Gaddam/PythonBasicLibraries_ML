```python
import seaborn as sb
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tips = sb.load_dataset('tips')
```

```python
tips.head()
```

```python
# distribution plot with single column 
sb.distplot(tips['total_bill'])  # blue line indicates kernal density 
```

```python

# we can disable 
sb.distplot(tips['total_bill'],kde=False) 
```

```python
#
sb.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
sb.jointplot(x='total_bill',y='tip',data=tips)
# default plot is scattered plot 
```

```python
sb.jointplot(x='total_bill',y='tip',data=tips,kind='hexa')
# Hexagon plot is where major density ones are grouped together
```

```python
sb.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
# Regression plot is where to fit the distribution where maximum values exits
```

```python
sb.pairplot(tips)
# all columns with size where as last one is size vs size
```

```python
sb.pairplot(tips,hue='sex')  # categorical 
# all columns with size where as last one is size vs size
```

```python
sb.rugplot(tips['total_bill'])
# single line along the distribution 
```

```python
sb.distplot(tips['total_bill'],kde=False)
```

```python
# kernal distribution esimation sum of all normal distribution
# Normal or Guasian distribution where we see how people do in the test 
# or Ranking in employee system
sb.distplot(tips['total_bill'],kde=True)
```

```python
sb.kdeplot(tips['total_bill'])
```

```python
g = sb.FacetGrid(tips, col="time",  row="sex")
g.map_dataframe(sb.histplot, x="total_bill")
g.set_axis_labels("Total bill", "Count")
```

```python
g = sb.FacetGrid(tips, col="time",  row="sex")
g.map_dataframe(plt.hist, x="total_bill")
g.set_axis_labels("Total bill", "Count")
```
